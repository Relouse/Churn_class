Видим, что пользователи, которые начали использовать продукт 11 месяцев назад, совершают самое большое кол-во звонков

С увеличением времени эксплуатирования, количество звонков уменьшается

Можно предположить, что 11 месяцев назад произошел всплеск новых пользователей

```python
df = pd.merge(df, target.to_frame(),left_index=True, right_index=True) 
month_totcalls = df.groupby('months').agg({'totcalls' : 'sum'}).reset_index()
sns.relplot(data=month_totcalls, x="months", y="totcalls")
```

```python
month_totpersons = df.groupby('months').agg({'months' : 'count'}).rename(columns={'months' : 'totpersons'}).reset_index()
sns.relplot(data=month_totpersons, x="months", y="totpersons")
```

Видим, что распределение звонков очень схоже с распределение общего количества пользователей по месяцам

Отсюда, можно сделать вывод, что гипотеза оправдана

## Посмотрим распределение количество звонков на человека по месяцам

```python
#churn == 0
m_tc_pers = df[df['churn'] == 0].groupby('months')\
             .agg({'totcalls' : 'sum', 'months' : 'count'})\
             .rename(columns={'months' : 'totpersons'})\

m_tc_pers['totcalls_per_pers'] = m_tc_pers['totcalls']/m_tc_pers['totpersons']

sns.relplot(data=m_tc_pers, x="months", y="totcalls_per_pers")


#churn == 1
m_tc_pers = df[df['churn'] == 1].groupby('months')\
             .agg({'totcalls' : 'sum', 'months' : 'count'})\
             .rename(columns={'months' : 'totpersons'})\

m_tc_pers['totcalls_per_pers'] = m_tc_pers['totcalls']/m_tc_pers['totpersons']

sns.relplot(data=m_tc_pers, x="months", y="totcalls_per_pers")
```

# Видим интересную закономерность - до определенного момента количество звонков на пользователя имеет практически линейную зависимость со временеи пользования

Теперь посмотрим, как количество звонков на человека взаимосвязано с процентом оттока пользователей

Для этого сгруппируем по месяцу и посчитаем количество звонков и отток



```python
df[['area','months','ethnic']] = df[['area','months','ethnic']].astype(int)
m_tc_ch = df.groupby('months')\
             .agg({'totcalls' : 'sum', 'churn' : 'sum', 'months' : 'count'})\
             .rename(columns={'months' : 'totpersons'})\

m_tc_ch['totcalls_per_pers'] = m_tc_ch['totcalls']/m_tc_ch['totpersons']
m_tc_ch['churn_per'] = m_tc_ch['churn']/m_tc_ch['totpersons']
sns.relplot(data=m_tc_ch, x="churn_per", y="totcalls_per_pers")
```

```python
m_tc_ch = m_tc_ch.drop(['totcalls', 'churn','totpersons'], axis=1)
m_tc_ch['churn_per'] = m_tc_ch['churn_per']*10 # Умножим отношение оттока на 10 для наглядности
m_tc_ch.plot()
```

#**Сравнение алгоритмов**

```python
X = df.drop('churn', axis=1)
y = df['churn']
```

```python
seed = 42

models = []
models.append(('LR', LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('DT', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('Random Forest', RandomForestClassifier()))
models.append(('XGBoost',XGBClassifier()))
models.append(('ADABoost',AdaBoostClassifier()))

print("Algorithm: Average Accuracy (Average Standard deviation) ")
results = []
names = []
scoring = 'accuracy'
for name, model in models:       
    start_time = time.time()
    kfold = model_selection.KFold(n_splits=10, random_state=seed,shuffle=True)
    
    
    cv_results = model_selection.cross_val_score(model, X, y, cv=kfold, scoring=scoring)        
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg,"Processing time",":{0:.2f} seconds".format(time.time() - start_time))

# boxplot
fig = plt.figure(figsize=(20,10))
fig.suptitle('Сравнение алгоритмов')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```

#Детально разберем каждый из алгоритмов

```python
def print_score(clf, X_train, X_test, y_train, y_test, train=True):
    
    lb = preprocessing.LabelBinarizer()
    lb.fit(y_train)
    if train:
        '''
        training performance
        '''
        res_train = clf.predict(X_train)
        print("Train Result:\n")
        print("accuracy score: {0:.4f}\n".format(accuracy_score(y_train, 
                                                                res_train)))
        print("Classification Report: \n {}\n".format(classification_report(y_train, 
                                                                            res_train)))
        print("Confusion Matrix: \n {}\n".format(confusion_matrix(y_train, 
                                                                  res_train)))
        print("ROC AUC: {0:.4f}\n".format(roc_auc_score(lb.transform(y_train), 
                                                      lb.transform(res_train))))

        res_cross = cross_val_score(clf, X_train, y_train, cv=10, scoring='accuracy')
        print("The cross validation score are: ","\n", res_cross,"\n")
        print("Average Accuracy: \t {0:.4f}".format(np.mean(res_cross)))
        print("Accuracy SD: \t\t {0:.4f}".format(np.std(res_cross)))
        
    elif train==False:
        '''
        test performance
        '''
        res_test = clf.predict(X_test)
        print("Test Result:\n")        
        print("accuracy score: {0:.4f}\n".format(accuracy_score(y_test, 
                                                                res_test)))
        print("Classification Report: \n {}\n".format(classification_report(y_test, 
                                                                            res_test)))
        print("Confusion Matrix: \n {}\n".format(confusion_matrix(y_test, 
                                                                  res_test)))   
        print("ROC AUC: {0:.4f}\n".format(roc_auc_score(lb.transform(y_test), 
                                                      lb.transform(res_test))))
```
